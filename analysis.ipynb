```python
def get_region(country):
    europe = ['Switzerland', 'Norway', 'Iceland', 'Denmark', 'Ireland', 'Luxembourg',
              'Sweden', 'Finland', 'United Kingdom', 'Netherlands', 'Austria', 'Belgium',
              'Germany', 'France', 'Italy', 'Spain', 'Portugal', 'Greece']

    north_america = ['United States', 'Canada', 'Bermuda', 'Bahamas']

    asia = ['Singapore', 'Japan', 'South Korea', 'Hong Kong', 'Israel', 'United Arab Emirates',
            'Qatar', 'Kuwait', 'China', 'India', 'Pakistan', 'Thailand', 'Malaysia']
    if country in europe:
        return 'Europe'
    elif country in north_america:
        return 'North America'
    elif country in asia:
        return 'Asia'
    else:
        return 'Other'

df['Region'] = df['Country'].apply(get_region)
```

```python
plt.figure(figsize=(10, 6))
region_avg = df.groupby('Region')['Cost of Living Index'].mean().sort_values(ascending=False)
plt.bar(region_avg.index, region_avg.values, color=['red', 'blue', 'green', 'orange'])
plt.xlabel('Region')
plt.ylabel('Average Cost of Living Index')
plt.title('Average Cost of Living Index by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\n=== CORRELATION ANALYSIS ===")
correlation_matrix = df[['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index',
                        'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Cost of Living Indicators')
plt.tight_layout()
plt.show()
```

```python
df['Value_Score'] = df['Local Purchasing Power Index'] / df['Cost of Living Index']
best_value = df.nlargest(10, 'Value_Score')[['Country', 'Value_Score', 'Local Purchasing Power Index', 'Cost of Living Index']]
print("\n=== TOP 10 BEST VALUE COUNTRIES ===")
print(best_value)
```

```python
worst_value = df.nsmallest(10, 'Value_Score')[['Country', 'Value_Score', 'Local Purchasing Power Index', 'Cost of Living Index']]
print("\n=== TOP 10 WORST VALUE COUNTRIES ===")
print(worst_value)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Best value countries
best_10 = best_value.head(10)
ax1.barh(best_10['Country'], best_10['Value_Score'], color='lightgreen')
ax1.set_title('Top 10 Best Value Countries\n(High Purchasing Power vs Cost)')
ax1.set_xlabel('Value Score (Purchasing Power / Cost)')
# Worst value countries
worst_10 = worst_value.head(10)
ax2.barh(worst_10['Country'], worst_10['Value_Score'], color='lightcoral')
ax2.set_title('Top 10 Worst Value Countries\n(Low Purchasing Power vs Cost)')
ax2.set_xlabel('Value Score (Purchasing Power / Cost)')

plt.tight_layout()
plt.show()
```

```python
df['Rent_Affordability'] = df['Local Purchasing Power Index'] / df['Rent Index']
most_affordable_rent = df.nlargest(10, 'Rent_Affordability')[['Country', 'Rent_Affordability', 'Local Purchasing Power Index', 'Rent Index']]
least_affordable_rent = df.nsmallest(10, 'Rent_Affordability')[['Country', 'Rent_Affordability', 'Local Purchasing Power Index', 'Rent Index']]

print("\n=== TOP 10 MOST RENT AFFORDABLE COUNTRIES ===")
print(most_affordable_rent)

print("\n=== TOP 10 LEAST RENT AFFORDABLE COUNTRIES ===")
print(least_affordable_rent)
```
